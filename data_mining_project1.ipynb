```python
import pandas as pd

match_data_raw = pd.read_csv('match_data.csv')
match_participants_raw = pd.read_csv('match_participants.csv')
summoners_raw = pd.read_csv('summoners.csv')


```

### adding columns to data_participants [winner_team duration iswin tier rank lp wins losses]

```python


match_participants = match_participants_raw.merge(match_data_raw[['match_id', 'winner_team','duration']], on='match_id', how='left')


match_participants['iswin'] = match_participants.apply(lambda row: 1 if row['team'] == row['winner_team'] else 0, axis=1)
match_participants.head(20)
```

```python
match_participants = match_participants.merge(summoners_raw[['puuid','tier','rank','lp','wins','losses']], on='puuid', how='left')
match_participants.head(20)
```

### Deleteing useless columns

```python
match_participants.drop(columns=['Unnamed: 0' ,'match_id', 'puuid', 'team', 'winner_team'], inplace=True)
match_participants.head(20)
```

### Finding missing values

```python
print(match_participants.isnull().sum())


match_participants.drop(match_participants[match_participants['role'] == 'Invalid'].index, inplace=True)
```

### Missing values in tier are due to riotAPI returning null for unranked players. We could either delete these rows or flag them as unranked and assume 50% win rate. The second option is a bit more risky so for simplicity let's choose the first one. The invalid part is about api scraping most likely collecting wrong gamemode or player behaving in an unpredictable way.

```python
match_participants.dropna(inplace=True)
```

### Let's change cs and vision score metrics to cs/min and vs/min

```python
match_participants['duration_minutes'] = match_participants['duration'] / 60

match_participants['cs_per_min'] = match_participants['cs'] / match_participants['duration_minutes']
match_participants['vs_per_min'] = match_participants['vision_score'] / match_participants['duration_minutes']

match_participants.drop(columns=['duration_minutes','cs','vision_score','duration'], inplace=True)

match_participants.head(20)
```

### Merge tier rank and lp into lp only

```python
tier_order = {
    'IRON': 0,
    'BRONZE': 1,
    'SILVER': 2,
    'GOLD': 3,
    'PLATINUM': 4,
    'EMERALD': 5,
    'DIAMOND': 6,
    'MASTER': 7,
    'GRANDMASTER': 8,
    'CHALLENGER': 9
}

rank_order = {
    'IV': 0,
    'III': 1,
    'II': 2,
    'I': 3
}

match_participants['tier_value'] = match_participants['tier'].map(tier_order)
match_participants['rank_value'] = match_participants['rank'].map(rank_order)


match_participants['total_lp'] = (match_participants['tier_value'] * 400 + 
                                 match_participants['rank_value'] * 100 + 
                                 match_participants['lp'])

match_participants.drop(columns=['tier', 'rank', 'lp', 'tier_value', 'rank_value'], inplace=True)

match_participants.head()
```

### Change wins and losses into match_count and winrate

```python
match_participants['match_count'] = match_participants['wins'] + match_participants['losses']
match_participants['winrate'] = match_participants['wins'] / match_participants['match_count']

match_participants.drop(columns=['wins', 'losses'], inplace=True)

match_participants.head()
```

### Let's detect outliers (cs/min,vs/min and assists aren't considered due to differences between supports and rest of the roles)

```python
def find_outliers(df, column):
    Q1 = df[column].quantile(0.2)
    Q9 = df[column].quantile(0.8)
    IQR = Q9 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q9 + 1.5 * IQR
    outliers  = (df[column] < lower_bound) | (df[column] > upper_bound)
    return outliers

kills_outliers = find_outliers(match_participants, 'kills')
deaths_outliers = find_outliers(match_participants, 'deaths')
gold_outliers = find_outliers(match_participants, 'gold_earned')
damage_outliers = find_outliers(match_participants, 'damage_dealt')
match_count_outliers = find_outliers(match_participants, 'match_count')

outliers = (kills_outliers | deaths_outliers | gold_outliers | damage_outliers | match_count_outliers)

match_participants = match_participants[~outliers]

# dectecting smurfs and boosted accounts
unusual_win_rate_condition = (
    ((match_participants['winrate'] < 0.35) | (match_participants['winrate'] > 0.70)) & 
    (match_participants['match_count'] > 50)
)

match_participants = match_participants[~unusual_win_rate_condition]
match_participants.head(10)

```

### Let's normalize the values now

```python
from sklearn.preprocessing import MinMaxScaler


numerical_columns = ['kills', 'deaths', 'assists', 'damage_dealt', 'gold_earned', 
                     'cs_per_min', 'vs_per_min', 'total_lp', 'match_count']

scaler = MinMaxScaler()

match_participants[numerical_columns] = scaler.fit_transform(match_participants[numerical_columns])

match_participants.head()
```

### Let's perform one hot encoding for champions and roles

```python
champion_dummies = pd.get_dummies(match_participants['champion'], prefix='champion')
role_dummies = pd.get_dummies(match_participants['role'], prefix='role')

match_participants = pd.concat([match_participants, champion_dummies, role_dummies], axis=1)

match_participants.drop(columns=['champion', 'role'], inplace=True)

match_participants.head()
```

```python
match_participants['is_win'] = match_participants['iswin']
match_participants.drop(columns=['iswin'], inplace=True)
match_participants.head(20)
```

```python
match_participants.to_csv('match_participants_updated.csv', index=False)
```
