```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
```

```python
df = pd.read_csv('match_participants_updated.csv')
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
sns.histplot(df['kills'], kde=True, color="blue", label="Kills", bins=20)
plt.title("Distribution of Kills")
plt.xlabel("Normalized Kills")

plt.subplot(2, 2, 2)
sns.histplot(df['deaths'], kde=True, color="red", label="Deaths", bins=14)
plt.title("Distribution of Deaths")
plt.xlabel("Normalized Deaths")

plt.subplot(2, 2, 3)
sns.histplot(df['assists'], kde=True, color="green", label="Assists", bins=20)
plt.title("Distribution of Assists")
plt.xlabel("Normalized Assists")

plt.tight_layout()
plt.show()
```

```python
numeric_features = ['kills', 'deaths', 'assists', 'damage_dealt', 'gold_earned', 
                    'cs_per_min', 'vs_per_min', 'total_lp', 'match_count', 'winrate']

corr_matrix = df[numeric_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Selected Features")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='kills', y='gold_earned', hue='is_win', palette={0: 'red', 1: 'green'}, alpha=0.6)
plt.title("Kills vs. Gold Earned by Match Outcome")
plt.xlabel("Normalized Kills")
plt.ylabel("Normalized Gold Earned")
plt.legend(title="Win")
plt.show()
```
